```python
import pandas as pd
```

```python
officer_info=pd.read_csv('Chicago_Data/roster_1936-2017_2017-04.csv',index_col='row_id')
officer_info
```

```python
unit_ref=pd.read_csv('Chicago_Data/unit_reference.csv')#,index_col='current_unit')
unit_ref
```

```python
unit_dict=unit_ref[['current_unit_description (April 2017 roster)','current_unit']].to_dict()
district_dict=(unit_ref[unit_ref['current_unit']<26][['current_unit_description (April 2017 roster)']]).dropna().to_dict()
district_map=district_dict['current_unit_description (April 2017 roster)']
district_map
```

```python
district_map = {value:key for key, value in district_map.items()}
district_map
```

```python
officer_info['assigned_district']=officer_info['unit_description'].map(district_map)
```

```python
officer_info.columns
```

```python
officer_current=officer_info[officer_info['current_status']==1]
```

```python
officer_current[['current_rank','assigned_district','first_name']].dropna().groupby(['assigned_district','current_rank']).count()

```

```python
police_officer_current=officer_current[officer_current['current_rank']=='POLICE OFFICER']
district_staffing=police_officer_current[['current_rank','assigned_district']].dropna().groupby(['assigned_district']).count()
district_staffing.columns=['#Officers']
```

```python
district_staffing.describe()
```

```python
police_beats=pd.read_csv('Simulation_Data/PoliceBeat.csv')
```

```python
police_beats
```

```python
district_to_beat=police_beats[['DISTRICT','BEAT']].groupby(['DISTRICT']).count()
district_to_beat.columns=['#Beats']
```

```python
beat_staffing=district_to_beat.merge(district_staffing,how='left',left_index=True,right_on='assigned_district')
beat_staffing.dropna(inplace=True)
beat_staffing['officers/beat']=round(beat_staffing['#Officers']/beat_staffing['#Beats'])
```

```python
beat_staffing['per_shift']=round(beat_staffing['officers/beat']/4) #4.2 ---> 5 when you fator in sick lave 
```
